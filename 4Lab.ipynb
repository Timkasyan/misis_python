1) Средний бал и список отстающих учеников

```python
Иванов О. 4
Петров И. 3
Дмитриев Н. 2
Смирнова О. 4
Керченских В. 5
Котов Д. 2
Бирюкова Н. 1
Данилов П. 3
Аранских В. 5
Лемонов Ю. 2
Олегова К. 4
```

```python
with open('Lab4.1.txt', 'r', encoding='utf-8') as f:
    summ = 0
    count = 0
    stroka = f.readline()
    while stroka != "":
        sList = stroka.split()
        summ += int(sList[-1])
        count += 1
        stroka = f.readline()
    if count != 0:
        print(f'Среднее: {round(summ/count,2)}')
```

2) Посчитать количество определенных слов в файле

```python
Понедельник
	Физика (лекц.)
	Физика (лаб.)
	Алгебра (практ.)
 
Вторник
	Геометрия (лекц.)
	Физика (практ.)
	Физика (лаб.)
	Физкультура (практ.)
...
```

```python
with open('Lab4.2.txt', 'r', encoding='utf-8') as f:
    lec = 0
    pract = 0
    lab = 0
    stroka = f.readlines()
    for i in stroka:
        if i.find('(лекц.)') != -1:
            lec +=1
        if i.find('(практ.)') != -1:
            pract +=1
        if i.find('(лаб.)') != -1:
            lab +=1

    print('Лекций:', lec)
    print('Практических:', pract)
    print('Лабораторных:', lab)
```

5) Напишите программу, которая принимает текст и выводит два слова: наиболее часто встречающееся и самое длинное.

```python
import collections

with open('Lab4.5.txt', 'r', encoding='utf-8') as f:
    text = f.read()
    words = text.split()
    counter = collections.Counter(words)
    most_common = counter.most_common()[0][0]
    longest = max(words, key=len)
    print(most_common, longest)
```
