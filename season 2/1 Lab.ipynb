## 1 Задание

```python
class Node:

    def __init__(self, data, next=None):
        self.data = data
        self.next = next

    def __str__(self): 
        if(self.next != None):
            return f"[Node with value {self.data}] \n{self.next}"
        else:
            return f"[Node with value {self.data}]"
        
class LinkedList:
    def __init__(self):
        self.head = None
    
    def __str__(self):
        return str(self.head)
```

```python
def reverse_list(head, tail=None):
    while head:
        head.next, tail, head = tail, head, head.next
    return tail
        
        
        
linked_list = LinkedList()
temp = Node(1)
linked_list.head = temp
for i in range(2, 6):
    temp.next = Node(i)
    temp = temp.next

print(linked_list,"\n---")

h = reverse_list(linked_list.head)

print(h)
```

```python
import random

def sortLinkedList(mainHead):
    head = mainHead
    while head:
        tempHead = head.next
        while tempHead:
            if (head.data>tempHead.data):
                head.data, tempHead.data = tempHead.data, head.data
            tempHead = tempHead.next
        head = head.next
    return mainHead

linked_list = LinkedList()
temp = Node(1)
linked_list.head = temp
for i in range(2, 11):
    temp.next = Node(i+random.randint(0,50))
    temp = temp.next

print(linked_list,"\n---")

print(sortLinkedList(linked_list.head))
```
