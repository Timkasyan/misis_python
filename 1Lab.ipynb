1)Сделайте так, чтобы число секунд отображалось в виде дни:часы:минуты:секунды.

```python
seconds=int(input('Введите количество секунд '))
if seconds >= 60:
    min=seconds//60
    seconds = seconds - min*60
    if min >= 60:
        hours = min//60
        min = min - hours*60
        if hours >=24:
            days = hours//24
            hours = hours - days*24
            print(f'{days} дней ')
        print(f'{hours} часов ')
    print(f'{min} минут ')
print(f'{seconds} секунд')
```

2) Напишите калькулятор который запрашивает на входе две переменные и знак, и в соответствии с знаком ( + - * / ) выводит результат

```python
t1 = int(input('Введите первое число '))
t2 = int(input('Введите второе число '))
symbol = input('Введите знак ')
if symbol == '+':
    print(t1+t2)
elif symbol == '-':
    print(t1-t2)
elif symbol == '*':
    print(t1*t2)
elif symbol == '/':
    print(t1/t2)
```
